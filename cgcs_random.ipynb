```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
sensor_count=2000
coord=[]
# random.seed(1)
count=0
for i in range(500//11):
    for j in range(500//11):
        count+=1
        if(random.random()>0.2):
            temp=[random.randint(1,10) for _ in range(2)]
        else:
            temp=[random.randint(0,11) for _ in range(2)]
        coord.append([(temp[0]+(11*i))%500,(temp[1]+(11*j))%500])
        if(count==2000):
            break
arr=np.array(coord)
print(arr.min(axis=0),arr.max(axis=0))
print(len(coord))
```
